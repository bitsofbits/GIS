**Animating Rain Gauge Data**

Maricopa county has an extensive network of rain gauge data available [online](http://www.fcd.maricopa.gov/Rainfall/Raininfo/raininfo.aspx). The data can be interesting,
on those occasions when we have a good storm, however the the maps they present don't give a very good idea of the progress of a storm.  I recently decided I should familiarize myself with some of Python's tools for mapping, and presenting this rain gauge data in a more graphival seemed like it might be an interesting project.

This (of course) turned out to be a deeper rabbit hole than I expected, leading to a plethora of side topics. Some interesting, others less so. But those are a topic for another day. For now, on with the show:

```python
from __future__ import print_function
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.basemap import Basemap
```

**Drawing the Map**

The first step is to draw a background map. It turns out that the basemap package is perfect for this sort of thing. It not only will draw country borders, state and county borders (the latter two US only though), it will take care of transforming coordinates from latitude and longitude to map coordinates. 

```python
from collections import namedtuple

# Since we pass around the figure, axes, and basemap instances a lot, we wrap them up in a tuple
MapData = namedtuple('MapData', ["fig", "axes", "bmap"])

def draw_arizona(min_lon=-115.0, max_lon=-109.0, min_lat=30.0, max_lat=38.0, figsize=(6,6)):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0.1,0.1,0.8,0.8])
    m = Basemap(projection='merc', llcrnrlat=min_lat, urcrnrlat=max_lat, llcrnrlon=min_lon, urcrnrlon=max_lon, resolution='c')
    m.shadedrelief(scale=0.25)
    m.drawcountries()
    m.drawstates()
    m.drawcounties(linewidth=0.3)
    return MapData(fig, ax, m)

azmap = draw_arizona()
```

However, I'm not interested in drawing the entire state of Arizona since I only have rainfall data for the Phoenix metro area. So we zoom in tighter on the map:

```python
phxmap = draw_arizona( -113.75, -111, 32.25, 35.0)
```

**Drawing Roads**

I found a shapefile containing openstreetmap data for Arizona on mapcruzin.com.  The *fiona* package makes loading the shapefile easy and then it's mostly just a matter of figuring out what to plot, since the map data contains all roads in AZ and we only want major, landmark roads. After a some experimenting, I decided that plotting just the roads labelled 'motorway' was sufficient.

```python
import fiona
from matplotlib.path import Path
from matplotlib.collections import PathCollection

def draw_roads(themap, types=['motorway']):
    # This might be doable with basemap.readshapefile, but this seems cleaner
    # Shapefile downloaded from www.openstreetmap.org and www.mapcruzin.com 
    # (http://www.mapcruzin.com/download-shapefile/us/arizona_highway.zip)
    with fiona.open('shapefiles/arizona_highway/arizona_highway.shp', 'r') as source:
        types = set(types)
        road_coords = []
        for road in source:
            if road['properties']['TYPE'] not in types:
                continue
            geom = road['geometry']
            assert geom['type'] == 'LineString'
            # A LineString geometry is a list of (x, y) coordinate values.
            coords = geom['coordinates']
            lons, lats = np.array(coords).T
            # lons and lats are in degrees, but we need them them in plot coordinates. Conveniently, 
            # passing lat/lon to basemap instanace results in a set of plot coordinates.
            transformed = np.array(themap.bmap(lons, lats)).T
            # Use the plot coordinates to create a matplotlib Path instance, then store that away for future use.
            road_coords.append(Path(transformed))
        # Create a PatchCollection from the road Paths and plot it. 
        themap.axes.add_collection(PathCollection(road_coords, facecolors='none', edgecolors="brown", linewidth=1))
    
draw_roads(phxmap)

# Force the new plot to display
phxmap.fig
```

**Adding Rainfall data**

The next step is to grab some rainfall data. I'll include data here so you don't have to abuse the Maricopa County Flood Control District's servers. The first thing we need is the location of all the monitoring stations. We grab the latitude, longitude and id of all the precipitation measuring stations from a copy of `ALERT_sensors_all_by_id.csv`, the original of which can be found [here](http://www.fcd.maricopa.gov/Rainfall/metaData.aspx). For each station, we load the rainfall from a 72 hour period from 12:00 in the morning on 9/7/2014 to 12 midnight on 9/9/2014. The data was downloaded from the Maricopa Flood Control district's [Single-Sensor Data Report Generator](http://www.fcd.maricopa.gov/Rainfall/ALERT/ssdata.aspx) and is stored in text files with the name "station_{id}.txt". All of these files are packed in a zip file name `rainfall_data.zip` to reduce  clutter.

First we define a class that processes that data into something usefull and hold onto the results:

```python
from datetime import datetime, timedelta

def datetime_range(start, stop, step=timedelta(minutes=1)):
    """Analogous to the range function, but for datetime objects"""
    times = []
    while start < stop:
        times.append(start)
        start = start + step
    return times

class Station(object):
    
    def __init__(self, station_id, raw_lat, raw_lon, raw_data):
        self.station_id = station_id
        # Longitude is west so we invert sign.
        self.lon = -self._coord_to_float(raw_lon)
        self.lat = self._coord_to_float(raw_lat)
        self.data = self.cook(raw_data)
        
    def hourly_rainfall(self, time):
        times = datetime_range(time - timedelta(minutes=60), time, timedelta(minutes=1))
        total = 0.0
        for t in times:
            total += self.data.get(t, 0.0)
        return total
                
    @staticmethod
    def cook(raw_data):
        # Rain data consists of a preamble of 6 lines, then a data in Month/Day/Year format followed by lines in
        # "time precip" where time is in military time format (eg 2215 for 10:15 PM).
        date = None
        cooked = {}
        # Skip preamble by starting after line 6.
        for line in raw_data[6:]:
            if '/' in line:
                month, day, year = (int(x) for x in line.split('/'))
                date = (month, day, year)
            else:
                if date is None:
                    raise ValueError("date not set before time-precip data encountered")
                time, precip = line.split()
                stamp = (year, month, day, int(time))
                if precip == "_____":
                    precip = 0  # use 0 for unknown? data
                cooked[Station._to_datetime(stamp)] = float(precip)
        return cooked
    
    @staticmethod
    def _to_datetime(ts):
        year, month, day, time24 = ts
        year += 2000
        hour = time24 // 100
        minute = time24 % 100
        if hour == 24:
            hour = 0
            return datetime(year, month, day, hour, minute) + timedelta(days=1)         
        else:
            return datetime(year, month, day, hour, minute)
    
    @staticmethod
    def _coord_to_float(text):
        degrees, minutes, seconds = (float(x) for x in text.split())
        return degrees + minutes / 60.0 + seconds / 3600       
```

Then we simply loop through all of the IDs in the weather station data and load the data for every station that measures
precipitation (as opposed to, for instance, stream flow).

```python
import csv
from collections import OrderedDict
from zipfile import ZipFile

station_data = OrderedDict()
with ZipFile("rainfall_data.zip") as rainfall_data:
    with rainfall_data.open('rainfall_data/ALERT_sensors_all_by_id.csv') as csvfile:
        csv_data = csv.reader(csvfile)
        next(csv_data) # discard header
        for (raw_id, name, type_, date, raw_lat, raw_lon, elev, locstring) in csv_data:
            if type_ == "Precip.":
                # For some reason, station ID is represented as "{ID}.00", so we discard the unused trailing part here. 
                station_id = int(raw_id.split('.')[0])
                with rainfall_data.open("rainfall_data/station_{0}.txt".format(station_id)) as source:
                    raw_data = source.readlines()   
                station_data[station_id] = Station(station_id, raw_lat, raw_lon, raw_data)
```

Now that we have the data for each station, let's plot them on the map.

```python
from matplotlib.patches import Circle

def plot_weather_station_locs(themap):
    for rdata in station_data.values():
        x, y = themap.bmap(rdata.lon, rdata.lat)
        # We use a radius of 200 since the x, y coordinates are in meters. Smaller values result
        # in dots that are too small or invisible.
        p = Circle((x,y), radius=200, facecolor='black', edgecolor='black')
        themap.axes.add_patch(p)
    
plot_weather_station_locs(phxmap)
phxmap.fig
```

The final step in preparing the data is to convert it into hourly rainfall data rather than 5-minute rainfall data. This smoothes out the resultant animation, making it clearer and more pleasant to watch. The hour that the rainfall is integrated over is a sliding window, so we still update at five minute intervals.

```python
def normalized_hourly_rainfall(stat_data, times):
    max_rainfall = 0.0
    rainfall = {}
    for t in times:
        hourly = np.array([x.hourly_rainfall(t) for x in stat_data.values()])
        rainfall[t] = hourly
        max_rainfall = max(max_rainfall, hourly.max())
    # Normalize the rainfall to the max rainfall.
    for x in rainfall.values():
        x /= max_rainfall
    return rainfall

all_times = datetime_range(datetime(2014, 9, 7, 12, 0), datetime(2014, 9, 9, 12, 1))
rainfall = normalized_hourly_rainfall(station_data, all_times)
```

Now all we need to do is to define a function to animate the data. Fortunately matplotlib has some pretty powerful [animation
features](http://matplotlib.org/api/animation_api.html) built in. We'll use the FuncAnimation function. This allows
us to pass in a function that is used to update the plot at each time step. Here we place a circle with a radius proportional
to the amount of rain in the last hour at each station location and update it at each time step.

I experimented with some other ways to display the rainfall data, including interpolating it onto a contour map. However the result ended up with a fair amount of artifacts from the interpolation, particularly near the edge, and I wasn't very happy with it. 

```python
from matplotlib.patches import Circle
from matplotlib import animation
    
def animate_rainfall(themap, station_data, rainfall, times, scale=20000):
    # Make a black dot for each station *and* and paritally transparent blue circle.
    # The dot is used to represent the locations of the station while the area of the
    # circle is updated at each time step to indicate the amount of rainfall.
    patches = []
    #
    def init():
        global label
        for rdata in station_data.values():
            x, y = themap.bmap(rdata.lon, rdata.lat)
            p = Circle((x,y), radius=1, facecolor='blue', edgecolor='none', alpha=0.3)
            themap.axes.add_patch(p)
            patches.append(p)
        # Place a label in the lower-left corner of the map than displays the current time.
        x0 = themap.axes.get_xlim()[0]
        y0 = themap.axes.get_ylim()[0]
        label = themap.axes.text(x0, y0, "")
    #
    def animate(dt):
        print('.', end='')
        if dt in rainfall:
            for w, p in zip(rainfall[dt], patches):
                # We set the *area* of the circle to be proportional to the rainfall at this time.
                p.radius = scale*np.sqrt(w)
        # Update the time and date.
        label.set_text("{0.year}:{0.month:02}:{0.day:02}:{0.hour:02}{0.minute:02}".format(dt))
        return patches + [label]
    #
    return animation.FuncAnimation(themap.fig, animate, 
                                   init_func=init,
                                   frames=times, 
                                   interval=20,
                                   blit=True)
```

**Animation**

We now finally get to create the animation. All that is required is invoking the save method to write out the animation to the named file. The options below are necessary for display on using quicktime on OS X, but seem to work fine with Chrome as well. The animation function prints out some dots as it progresses just so that we know something is happening. This may take quite a while to run.

```python
anim = animate_rainfall(phxmap, station_data, rainfall, all_times)
# These options work for quicktime on OS X. It's possible that some other options may be required for other machines.
anim.save("September_8th_storm_rainfall.m4v", fps=60, extra_args=['-vcodec', 'libx264', '-pix_fmt','yuv420p'])
# This spits out a lot of dots as well as a still of the first frame of the video.
```

```python
# From http://nbviewer.ipython.org/github/fperez/ipynb-embed-video/blob/master/EmbeddedVideo.ipynb
def video(fname, mimetype):
    """Load the video in the file `fname`, with given mimetype, and display as HTML5 video.
    """
    from IPython.display import HTML
    video_encoded = open(fname, "rb").read().encode("base64")
    video_tag = '<video controls alt="test" src="data:video/{0};base64,{1}">'.format(mimetype, video_encoded)
    return HTML(data=video_tag)
```
